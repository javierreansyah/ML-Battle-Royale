```python
import pandas as pd
```

```python
df = pd.read_csv('../dataset/train.csv')
df.shape
```

```python
df = df.drop(['Name'], axis=1)
df.head()
```

```python
df = df.drop(df[df['Cabin'].isnull()].index)
df.shape
```

```python
df.isnull().sum()
```

```python
df[['GroupNumber', 'PassengerNumber']] = df['PassengerId'].str.split('_', expand=True)
df = df.drop(['PassengerId'], axis=1)
df = df.drop(['PassengerNumber'], axis=1)
df['GroupNumber'].astype(int)

df.head()
```

```python
df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'RoomService']] = df[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'RoomService']].fillna(value=0)
df[['HomePlanet', 'Destination']] = df[['HomePlanet', 'Destination']].fillna(value='Unknown')
df.isnull().sum().sort_values(ascending=False)
```

```python
df['Transported'] = df['Transported'].astype(int)
df['VIP'] = df['VIP'].astype(int)
df['CryoSleep'] = df['CryoSleep'].astype(int)

df.head()
```

```python
df[["Deck", "Cabin_num", "Side"]] = df["Cabin"].str.split("/", expand=True)
df = df.drop(['Cabin'], axis=1)
df = df.drop(['Cabin_num'], axis=1)
df_deck_side = df[['Deck', 'Side']]

df.head()
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5)
df['Age'] = imputer.fit_transform(df[['Age']])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output = False).set_output(transform='pandas')

ohetransform = ohe.fit_transform(df[['Deck', 'Destination','Side', 'HomePlanet']])
ohetransform

df = pd.concat([df, ohetransform], axis=1).drop(columns=['Deck', 'Destination','Side', 'HomePlanet'])
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Reset display options to default
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')

# Separate features (X) and target variable (y)
X = df.drop('Transported', axis=1)
y = df['Transported']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the scaled data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Display shapes of the train and test sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier

classifiers = {
    'Random Forest': {
        'classifier': RandomForestClassifier(),
        'param_grid': {
            'classifier__n_estimators': [10, 50, 100],
            'classifier__max_depth': [None, 10, 20, 30],
            'classifier__min_samples_split': [2, 5, 10]
        }
    },
    'SVM': {
        'classifier': SVC(),
        'param_grid': {
            'classifier__C': [0.1, 1, 10, 100],
            'classifier__gamma': ['scale', 'auto'],
            'classifier__kernel': ['linear', 'poly', 'rbf', 'sigmoid']
        }
    },
    'Logistic Regression': {
        'classifier': LogisticRegression(),
        'param_grid': {
            'classifier__C': [0.1, 1, 10, 100],
            'classifier__solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
        }
    },
    'KNN': {
        'classifier': KNeighborsClassifier(),
        'param_grid': {
            'classifier__n_neighbors': [3, 5, 7, 9, 11],
            'classifier__weights': ['uniform', 'distance'],
            'classifier__metric': ['euclidean', 'manhattan']
        }
    },
    'Decision Tree': {
        'classifier': DecisionTreeClassifier(),
        'param_grid': {
            'classifier__max_depth': [None, 10, 20, 30],
            'classifier__min_samples_split': [2, 5, 10]
        }
    },
    'Gradient Boosting': {
        'classifier': GradientBoostingClassifier(),
        'param_grid': {
            'classifier__n_estimators': [50, 100, 200],
            'classifier__learning_rate': [0.05, 0.1, 0.5],
            'classifier__max_depth': [3, 4, 5]
        }
    },
    'Naive Bayes': {
        'classifier': GaussianNB(),
        'param_grid': {}
    },
    'SGD Classifier': {
        'classifier': SGDClassifier(),
        'param_grid': {
            'classifier__loss': ['hinge', 'log', 'modified_huber', 'squared_hinge', 'perceptron'],
            'classifier__penalty': ['l2', 'l1', 'elasticnet'],
            'classifier__alpha': [0.0001, 0.001, 0.01, 0.1]
        }
    },
    'MLP Classifier': {
        'classifier': MLPClassifier(),
        'param_grid': {
            'classifier__hidden_layer_sizes': [(50,50,50), (50,100,50), (100,)],
            'classifier__activation': ['tanh', 'relu'],
            'classifier__solver': ['sgd', 'adam'],
            'classifier__alpha': [0.0001, 0.05],
            'classifier__learning_rate': ['constant','adaptive']
        }
    },
    'AdaBoost': {
        'classifier': AdaBoostClassifier(),
        'param_grid': {
            'classifier__n_estimators': [50, 100, 200],
            'classifier__learning_rate': [0.05, 0.1, 0.5]
        }
    }
}

for name, classifier in classifiers.items():
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', classifier['classifier'])
    ])
    
    param_grid = classifier['param_grid']
    
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    
    print(f"{name}:")
    print("Best Hyperparameters:", best_params)
    print("Best Accuracy:", best_score)
    print("------------------------------------------")
```
